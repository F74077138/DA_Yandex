### Предобработка данных

```python
old_data_size = data.shape[0]
# check missing values
data.isna().sum()
```

```python
# delete outlier
print(data['last_price'].min())
print(data['last_price'].max())
data['last_price'] = data['last_price'].replace(12190, 12190000)
data = data.drop(data[data['last_price'] > 500000000].index)
```

Самое минимальное значение в столбце равно 12190, что просто не может быть правдой. Скорее всего цена указана в тысячах. Также максимальное значение равно 763000000. Это значение может повлиять на результаты исследования, поэтому его следует удалить.

```python
# remove hidden duplicates from 'locality_name'
# delete rows with missing 'locality_name' values
data['locality_name'] = data['locality_name'].str.replace('ё', 'е')
data = data.drop(data[data['locality_name'].isna()].index)
```

Убираем неявные дубликаты заменой буквы **ё** на **е**.  Также удаляем строки, где отсутствует значение ‘locality_name’ (их всего 49).

```python
# fill missing 'floors_total' values
# change 'floors_total' type to integer
data['floors_total'] = data.apply(
    lambda row: row['floor'] if np.isnan(row['floors_total']) else row['floors_total'],
    axis=1
)
data['floors_total'] = data['floors_total'].astype('int')
```

Заменяем пропущенные значения в ‘floors_total’. Если значение отсутствует, то это скорее всего связано с тем, что жильё находится на последнем этаже (‘floor’). Этаж не может быть вещественным числом, поэтому заменяем тип на integer.

```python
# fill missing 'kitchen_area' and 'living_area' values with average coef. * total_area  
# remove rows where living_area + kitchen_area bigger than total_area
average_coef_kitchen = (data['kitchen_area'] / data['total_area']).mean()
average_coef_living = (data['living_area'] / data['total_area']).mean()

data['kitchen_area'] = data.apply(
    lambda row: average_coef_kitchen * row['total_area'] if np.isnan(row['kitchen_area']) else row['kitchen_area'],
    axis=1
)

data['living_area'] = data.apply(
    lambda row: average_coef_living * row['total_area'] if np.isnan(row['living_area']) else row['living_area'],
    axis=1
)

data = data.drop(data[(data['living_area'] + data['kitchen_area']) > data['total_area']].index)
```

Посчитаем средний коэффициент кухни/жилой площади на общую площадь и найдем ‘kitchen_area’ и ‘living_area’. Удалим значения, где сумма кухни и жилой площади превышает общую площадь, т.к. такие значения противоречат здравому смыслу.

```python
# set missing 'is_apartment' values to False
# change type to bool
data['is_apartment'] = data['is_apartment'].fillna(0)
data['is_apartment'] = data['is_apartment'].astype('bool')

# set missing 'balcony' to 0
# change type to integer
data['balcony'] = data['balcony'].fillna(0)
data['balcony'] = data['balcony'].astype('int')
```

Заполнить пустые значения в ‘is_apartment’ False (если пропуск, то скорее всего не апартаменты). Поменять тип на bool (значения True и False). 
Если значение в столбце ‘balcony’ отсутствует, то скорее всего балкон отсутствует в целом. Заменим тип на integer, потому что количество балконов не может быть вещественным числом.

```python
# fill missing 'days_exposition' with mean 
# change type to integer
#mean_value=data['days_exposition'].mean()
#data['days_exposition'].fillna(value=mean_value, inplace=True)
#data['days_exposition'] = data['days_exposition'].astype('int')
```

Заполним пропущенные значения в ‘days_exposition’ средним значением. Заменим тип на integer, потому что количество дней не может быть вещественным числом.

```python
# remove outliers from 'ceiling_height'
# fill missing values with mean
data.loc[(data['ceiling_height'] < 2.5) | (data['ceiling_height'] > 5), 'ceiling_height'] = None
mean_value = data['ceiling_height'].mean()
data['ceiling_height'].fillna(value=mean_value, inplace=True)
```

Полагаясь на здравый смысл, мы можем предположить, что потолки не могут быть выше 5 метров и ниже 2.5, поэтому мы заменим эти и пропущенные значения средним.

```python
# fill missing values 'parks_around3000' and 'ponds_around3000' (probably no park/pond in area)
# change type to integer
"""
data['parks_around3000'] = data['parks_around3000'].fillna(0.0)
data['ponds_around3000'] = data['ponds_around3000'].fillna(0.0)
data['parks_around3000'] = data['parks_around3000'].astype('int')
data['ponds_around3000'] = data['ponds_around3000'].astype('int')
"""
```

Если значение в ‘parks’ пропущено, то это может означать, что отсутствует парк/водоем в районе 3 км. Поэтому эти значения стоит заменить на 0. Заменим тип на integer, потому что количество парков и водоемов не может быть вещественным числом.

```python
# fill missing values 'parks_nearest' and 'ponds_nearest' (there is probably park/pond 3kw away)
"""
data['parks_nearest'] = data.apply(
    lambda row: 3001 if np.isnan(row['parks_nearest']) and row['parks_around3000']==0 else row['parks_nearest'],
    axis=1
)

data['ponds_nearest'] = data.apply(
    lambda row: 3001 if np.isnan(row['ponds_nearest']) and row['ponds_around3000']==0 else row['ponds_nearest'],
    axis=1
)
"""
```

Если нет парков/водоемов поблизости, это может означать, что они находятся дальше 3 км.

```python
# fill missing 'airports_nearest' and 'cityCenters_nearest' by calculating mean value in that area
"""
for d in data['locality_name'].unique():
    data.loc[(data['locality_name']==d) & (data['airports_nearest'].isna()), 'airports_nearest'] = data.loc[(data['locality_name']==d), 'airports_nearest'].mean()
    
for d in data['locality_name'].unique():
    data.loc[(data['locality_name']==d) & (data['cityCenters_nearest'].isna()), 'cityCenters_nearest'] = data.loc[(data['locality_name']==d), 'cityCenters_nearest'].mean()
"""
```

Заменить пропущенные значения в 'airports_nearest' и 'cityCenters_nearest' средним значением согласно расположению.

### Посчитайте и добавьте в таблицу новые столбцы

```python
# price for one sq. meter
data['sq_meter'] = round(data['last_price'] / data['total_area'], 2)

#day of the week/month/year
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')
data['weekday'] = data['first_day_exposition'].dt.weekday
data['month'] = data['first_day_exposition'].dt.month
data['year'] = data['first_day_exposition'].dt.year

# group by floor
data['floor_type'] = 'other'
data.loc[data['floor']==1, 'floor_type'] = 'first'
data.loc[data['floor']==data['floors_total'], 'floor_type'] = 'last'

# convert m to km 
data['cityCenters_km'] = round(data['cityCenters_nearest'] / 1000)

data.head(5)
```
